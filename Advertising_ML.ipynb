```python
def var_summary(x):
    return pd.Series([x.count(), 
        x.isnull().sum(),
        x.sum(),
        x.mean(),
        x.median(),
        x.std(),
        x.var(),
        x.min(),
        x.quantile(0.01),
        x.quantile(0.05),
        x.quantile(0.10),
        x.quantile(0.25),
        x.quantile(0.50),
        x.quantile(0.75),
        x.quantile(0.90),
        x.quantile(0.95),
        x.quantile(0.99),
        x.max()],
        index=['N',
               'NMISS',
               'SUM',
               'MEAN', 
               'MEDIAN', 
               'STD',
               'VAR', 
               'MIN', 
               'P1',
               'P5',
               'P10',
               'P25/Q1',
               'P50/Q2',
               'P75/Q3',
               'P90',
               'P95', 
               'P99',
               'MAX'])
```

```python
# let’s statistically confirm which advertising channels (TV, Radio, Newspaper) 
# ...significantly affect Sales using simple linear regression and then multiple regression.

import statsmodels.api as sm


# Simple Linear Regressions (one predictor at a time)

# Let’s check the impact of each channel individually on Sales:

#TV vs Sales

X = data_new['TV']
y = data_new['Sales']

X = sm.add_constant (X) # add intercept
model_tv = sm.OLS(y,X).fit()

print (model_tv.summary())


#Radio vs Sales

X = data_new['Radio']
y = data_new['Sales']

X = sm.add_constant (X) # add intercept
model_radio = sm.OLS(y,X).fit()

print (model_radio.summary())



#Newspaper vs Sales

X = data_new['Newspaper']
y = data_new['Sales']

X = sm.add_constant (X) # add intercept
model_newspaper = sm.OLS(y,X).fit()

print (model_newspaper.summary())


```

```python
# Now run the multiple regression (all three predictors together):

X = data_new[['TV', 'Radio', 'Newspaper']]
y = data_new['Sales']
X = sm.add_constant(X)
model_multi = sm.OLS(y, X).fit()
print(model_multi.summary())
```

```python
# Key Insights

# The model fits the data extremely well (R² = 0.897).

# TV and Radio together explain nearly all of the variation in Sales.

# Newspaper spending can be safely reduced or reallocated — it doesn’t improve Sales significantly once other media are accounted for.

# The negative sign on Newspaper (-0.001) likely results from multicollinearity (spending overlap) or noise — not a real negative effect.

# Final Regression Equation
# Sales = 2.939 + 0.0458(TV)+0.1885(Radio)−0.0010 (Newspaper )
# Sales=2.939+0.0458(TV)+0.1885(Radio)−0.0010(Newspaper)
```
